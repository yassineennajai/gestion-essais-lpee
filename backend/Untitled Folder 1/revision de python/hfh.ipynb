```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Charger les données
df = pd.read_csv('dm_office_sales.csv')

# 2. Créer le graphe Salary vs Sales avec la division
plt.figure(figsize=(12, 8))

# Utiliser seaborn pour un graphique plus esthétique avec coloration par division
scatter = sns.scatterplot(data=df, x='salary', y='sales', hue='division', 
                         palette='viridis', s=100, alpha=0.7)

# Personnaliser le graphique
plt.title('Salary vs Sales par Division', fontsize=16, fontweight='bold')
plt.xlabel('Salary', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.grid(True, alpha=0.3)

# Améliorer la légende
plt.legend(title='Division', title_fontsize=12, fontsize=10)

# Ajuster la mise en page
plt.tight_layout()

# 3. Exporter la figure en JPG
plt.savefig('salary_vs_sales_par_division.jpg', 
            dpi=300, 
            bbox_inches='tight', 
            format='jpg')

# Afficher le graphique
plt.show()

# Afficher quelques informations statistiques
print("Aperçu des données :")
print(df.head())
print(f"\nNombre total d'observations : {len(df)}")
print(f"Divisions présentes : {df['division'].unique()}")
```
