EXERCICE 2:

creer un dataframe

```python
import pandas as pd
import numpy as np

np.random.seed(101)
mydata = np.random.randint(0, 101, (4, 3))
myindex = ['CANADA', 'USA', 'FRANCE', 'UK']
mycolumns = ['Jan', 'Feb', 'Mar']

mydata = pd.DataFrame(mydata, index=myindex, columns=mycolumns)
print(mydata)
```

Exporter en CSV 

```python
mydata.to_csv('mydata.csv')
```

Afficher les informations et statistiques

```python
print(mydata.info())
print(mydata.describe())
```

Supprimer la ligne “FRANCE”

```python
mydata = mydata.drop('FRANCE')
```

Afficher uniquement les informations relatives à “USA”

```python
print(mydata.loc['USA'])
```

Donner la moyenne des valeurs du mois de janvier

```python
moyenne_janvier = mydata['Jan'].mean()
print(moyenne_janvier)
```
