Pourquoi la valeur minimale de MSE n’est pas atteinte ?

Parce que :

Le poids w1 est modifié de façon arbitraire (+0.4)

On ne calcule pas la dérivée de l’erreur

On ne suit pas la direction de descente

Donc il n’y a pas d’optimisation réelle.

3️Augmenter le nombre d’itérations garantit-il le minimum ?

 Non.

Même avec 10 000 itérations, si la règle de mise à jour est incorrecte, on ne convergera pas vers le minimum.

Le problème vient de la méthode, pas du nombre d’itérations.

 Les deux problèmes du code

Problème 1 :
Il n’y a pas de Gradient Descent (pas de calcul de dérivée).

 Problème 2 :
Il n’y a pas de learning rate (pas de contrôle de la taille du pas).

 Comment la dérivée partielle aide-t-elle ? (Technique 1)

La dérivée partielle permet de calculer le gradient :




Elle indique :

La direction dans laquelle modifier le paramètre

Si on doit augmenter ou diminuer w1

Donc elle résout :

 Le problème de la direction d’optimisation

 Comment le learning_rate aide-t-il ? (Technique 2)

La constante learning_rate contrôle la taille du pas :

w1=w1−α⋅gradient
w1=w1−α⋅gradient

Si le pas est :

Trop grand → on dépasse le minimum

Trop petit → convergence très lente



```python
import numpy as np

def mse(y_predicted, y_real):
    return np.mean((y_real - y_predicted) ** 2)

def learning2(X, y_real):
    w0 = 0
    w1 = 0
    learning_rate = 0.01
    nb_iterations = 200

    n = len(X)

    for i in range(nb_iterations):

        # prédiction
        y_predicted = w1 * X + w0

        # calcul de l’erreur
        error = mse(y_predicted, y_real)

        # calcul des dérivées partielles
        dw1 = (-2/n) * np.sum(X * (y_real - y_predicted))
        dw0 = (-2/n) * np.sum(y_real - y_predicted)

        # mise à jour des paramètres
        w1 = w1 - learning_rate * dw1
        w0 = w0 - learning_rate * dw0

        print("w0:{} ---- w1:{} ---- error:{}".format(dw0, dw1, error))
learning2(X,Y)
```

Exécuter le programme et analyser le MSE trouvé

Si on exécute le programme tel qu’il est donné (non corrigé) :

Le MSE ne converge pas vers le minimum.

Il peut rester presque constant ou même augmenter.

Parfois le programme génère une erreur car dw0 et dw1 ne sont pas définis.

 Conclusion : l’algorithme ne minimise pas correctement le MSE.

 Quel est le problème ? Dans quelles lignes ?

Il y a deux problèmes majeurs :

Problème 1 : le gradient n’est pas calculé



On met à jour w0 et w1 avec dw0 et dw1, mais ces dérivées ne sont jamais définies.

Problème 2 : mauvais signe dans la mise à jour

On utilise :



w=w+learning_rate⋅gradient



w=w+learning_rate⋅gradient




Alors qu’en descente de gradient on doit utiliser :

w=w−learning_rate⋅gradient
w=w−learning_rate⋅gradient

Sinon on fait une montée du gradient (Gradient Ascent).

Comment résoudre ce problème ?


Pour corriger l’algorithme, il faut calculer les gradients dw0 et dw1 à partir de l’erreur entre les valeurs réelles et prédites, puis mettre à jour les paramètres avec un signe négatif : w = w - learning_rate * gradient. Ces corrections permettent au MSE de diminuer progressivement et d’assurer la convergence vers une solution optimale.

```python


import numpy as np

def mse(y_predicted, y_real):
    return np.mean((y_real - y_predicted) ** 2)

def learning3(X, y_real):

    w0 = 0
    w1 = 0
    learning_rate = 0.001
    n = len(X)

    for i in range(1000):

        y_predicted = w1 * X + w0
        error = mse(y_predicted, y_real)

        dw1 = (-2/n) * np.sum(X * (y_real - y_predicted))
        dw0 = (-2/n) * np.sum(y_real - y_predicted)

        w0 = w0 - learning_rate * dw0
        w1 = w1 - learning_rate * dw1

        print("w0:{} -- w1:{} -- y_predected:{} --> ERROR: {}"
              .format(w0, w1, y_predicted, mse(y_predicted, y_real)))  # ligne 7

        w0 = w0 + learning_rate * w0   # ligne 8
        w1 = w1 + learning_rate * w1  
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

# Hyperparamètres
lr = 0.0001
n_iters = 1000

def fit(X, y):
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    bias = 0

    for _ in range(n_iters):
        y_predicted = np.dot(X, weights) + bias

        dw = (2/n_samples) * np.dot(X.T, (y_predicted - y))
        db = (2/n_samples) * np.sum(y_predicted - y)

        weights -= lr * dw
        bias -= lr * db

    return bias, weights

def predict(X, w, b):
    return np.dot(X, w) + b

# Chargement des données
df = pd.read_csv("advertising.csv")
y = df["Sales"].values
X = df[["TV", "Radio", "Newspaper"]].values

# Split train / test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=12
)

# Entraînement (descente de gradient)
b, w = fit(X_train, y_train)

# Prédictions
predictions = predict(X_test, w, b)
```

🔹 Pourquoi utiliser train_test_split ?

Pour séparer les données en :

```python
#Tracer la courbe du MSE
#🔹 Modification de la fonction fit
import numpy as np
import matplotlib.pyplot as plt

lr = 0.0001
n_iters = 1000

def fit(X, y):
    n_samples, n_features = X.shape
    weights = np.zeros(n_features)
    bias = 0
    cost_history = []

    for _ in range(n_iters):
        y_predicted = np.dot(X, weights) + bias
        
        # Calcul du MSE
        cost = np.mean((y_predicted - y) ** 2)
        cost_history.append(cost)

        # Gradient
        dw = (2/n_samples) * np.dot(X.T, (y_predicted - y))
        db = (2/n_samples) * np.sum(y_predicted - y)

        weights -= lr * dw
        bias -= lr * db

    return bias, weights, cost_history
```

```python
#Tracé de la courbe
b, w, cost_history = fit(X_train, y_train)

plt.plot(cost_history)
plt.xlabel("Itérations")
plt.ylabel("MSE")
plt.title("Évolution de la fonction coût")
plt.show()

#La courbe doit décroître progressivement vers le minimum.
```
