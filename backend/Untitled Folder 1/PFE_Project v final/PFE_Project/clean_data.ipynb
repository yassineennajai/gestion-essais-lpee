```python
dupl = df.duplicated(subset = ['sales', 'quantity', 'profit', 'orderdate_day', 'orderdate_month', 'orderdate_year'], keep=False)
df.loc[dupl]
```

```python
# here we deleted one of the 2 rows that have the same informations
df = df.drop(df[df['Unnamed: 0'] == 567 ].index)
```

```python
df.loc[dupl]
```

```python
# here we convert the weekday from german to english so it would be easy to combine all the date columns in one column
df['orderdate_weekday'] = df['orderdate_weekday'].replace({'Montag': 'Monday', 'Dienstag': 'Tuesday', 'Mittwoch': 'Wednesday', 'Donnerstag': 'Thursday', 'Freitag': 'Friday', 'Samstag': 'Saturday', 'Sonntag': 'Sunday'})
df['shipdate_weekday'] = df['shipdate_weekday'].replace({'Montag': 'Monday', 'Dienstag': 'Tuesday', 'Mittwoch': 'Wednesday', 'Donnerstag': 'Thursday', 'Freitag': 'Friday', 'Samstag': 'Saturday', 'Sonntag': 'Sunday'})
```

```python
df['order_date'] = pd.to_datetime(dict(day=df['orderdate_day'], month=df['orderdate_month'], year=df['orderdate_year']))
df['ship_date'] = pd.to_datetime(dict(day=df['shipdate_day'], month=df['shipdate_month'], year=df['shipdate_year']))
```

```python
df = df.drop(columns = ['orderdate_day', 'orderdate_month', 'orderdate_year', 'shipdate_year', 'shipdate_day', 'shipdate_month', 'Unnamed: 0'])
```
