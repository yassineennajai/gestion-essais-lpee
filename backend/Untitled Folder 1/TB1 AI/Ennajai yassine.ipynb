```python
##exercice 4:
import matplotlib.pyplot as plt


Periode = ['1 Mois','3 Mois','6 Mois','1 An','2 An','3 An','5 An','7 An','10 An','20 An','30 An']
juillet16_2007 = [4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
juillet16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]


plt.plot(Periode, juillet16_2007, marker='o', label='Juillet 16, 2007')
plt.plot(Periode, juillet16_2020, marker='s', label='Juillet 16, 2020')


plt.title("Courbes de rendement (2007 vs 2020)")
plt.xlabel("Période")
plt.ylabel("Rendement (%)")
plt.legend()
plt.grid(True)
plt.show()



```

Ce graphique montre une grande différence entre les deux périodes :

En 2007, l’économie était solide et les taux d’intérêt élevés.

En 2020, les taux sont très bas à cause des politiques monétaires pour relancer l’économie après la pandémi

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 8))

# Premier graphique : 2007
ax1.plot(Periode, juillet16_2007, color='blue', marker='o')
ax1.set_title("Courbe de rendement - Juillet 16, 2007")
ax1.set_ylabel("Rendement (%)")
ax1.grid(True)

# Deuxième graphique : 2020
ax2.plot(Periode, juillet16_2020, color='red', marker='s')
ax2.set_title("Courbe de rendement - Juillet 16, 2020")
ax2.set_xlabel("Période")
ax2.set_ylabel("Rendement (%)")
ax2.grid(True)

```

En 2007, la courbe de rendement est presque plate, traduisant des taux d’intérêt élevés et stables dans un contexte économique solide.
En revanche, en 2020, la courbe est ascendante avec des rendements très faibles à court terme, conséquence directe des politiques monétaires expansionnistes mises en place après la crise du COVID-19.

```python
fig, ax1 = plt.subplots(figsize=(8, 5))

# Axe principal (gauche) pour 2007
ax1.plot(Periode, juillet16_2007, color='blue', marker='o', label='2007')
ax1.set_xlabel("Période")
ax1.set_ylabel("Rendement 2007 (%)", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# Créer un deuxième axe Y (à droite)
ax2 = ax1.twinx()
ax2.plot(Periode, juillet16_2020, color='red', marker='s', label='2020')
ax2.set_ylabel("Rendement 2020 (%)", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Titre et affichage
plt.title("Comparaison des courbes de rendement (2007 vs 2020)")
fig.tight_layout()
plt.show()
```

Comparaison 2007 vs 2020

En 2007 : taux élevés → économie forte, inflation contrôlée.

En 2020 : taux très bas → politique monétaire expansionniste pour stimuler l’économie.

L’écart entre les deux courbes montre combien les conditions économiques et monétaires ont changé : de la stabilité pré-crise à une stimulation post-crise

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('dm_office_sales.csv')
print(df.head())  # Affiche les 5 premières lignes pour vérifier
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='salary', y='sales')
plt.title('Salary vs Sales')
plt.xlabel('Salary')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

Les employés ayant un salaire plus élevé réalisent souvent plus de ventes,
ou inversement, ceux qui font plus de ventes obtiennent un salaire plus élevé.
C’est typique dans les domaines où la rémunération dépend de la performance commerciale.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='salary', y='sales', hue='division', style='division', s=100)
plt.title('Salary vs Sales par Division')
plt.xlabel('Salary')
plt.ylabel('Sales')
plt.legend(title='Division')
plt.grid(True)
plt.show()
```

Globalement, il existe une forte relation positive entre le salaire et les ventes, quel que soit le département.

Cependant, certaines divisions montrent des performances moyennes ou extrêmes différentes.

Ce graphique est donc utile pour comparer la performance des divisions tout en observant la tendance globale de corrélation.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='salary', y='sales', hue='division', style='division', s=100)
plt.title('Salary vs Sales par Division')
plt.xlabel('Salary')
plt.ylabel('Sales')
plt.legend(title='Division')
plt.grid(True)
plt.savefig('salary_vs_sales_division_seaborn.jpg', dpi=300, bbox_inches='tight', format='jpg')
plt.show()
```
