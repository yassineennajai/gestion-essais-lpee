Exercice 1 : Bag of Words (BoW)
1️ Construction d’une représentation Bag of Words

Nous utilisons la classe CountVectorizer de la bibliothèque scikit-learn.

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "Fatima aime l IA",
    "Karim aime le NLP"
]

vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(documents)
```

2️ Vocabulaire et matrice BoW

```python
print("Vocabulaire :", vectorizer.get_feature_names_out())
print("Matrice BoW :")
print(X_bow.toarray())
```

3. Pourquoi l’ordre des mots n’est pas pris en compte ?

Le modèle Bag of Words considère un texte comme un ensemble de mots sans tenir compte de leur position.

Il ne garde que :

la présence du mot

ou son nombre d’occurrences

Ainsi, les phrases :

“Fatima aime l’IA”

“L’IA aime Fatima”

auront exactement la même représentation BoW, ce qui entraîne une perte du sens syntaxique.

Exercice 2 : TF-IDF

1️. Construction de la représentation TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(documents)
```

2️⃣ Vocabulaire et matrice TF-IDF

```python
print("Vocabulaire TF-IDF :", tfidf.get_feature_names_out())
print("Matrice TF-IDF :")
print(X_tfidf.toarray())
```

✔️ Les valeurs sont réelles (pondérations)
✔️ Plus un mot est rare, plus son poids est élevé

3️⃣ Comparaison BoW vs TF-IDF
Méthode	Principe
BoW	Compte brut des mots
TF-IDF	Importance relative des mots

🔹 Le mot “aime” apparaît dans toutes les phrases :

BoW → poids élevé

TF-IDF → poids faible

 Pourquoi TF-IDF réduit l’importance des mots fréquents ?

TF-IDF pénalise les mots présents dans beaucoup de documents, car ils sont peu discriminants.

Formule :

TF-IDF=TF×IDF
TF-IDF=TF×IDF

TF : fréquence du mot dans le document

IDF : inverse de la fréquence du mot dans le corpus

Exercice 3 : Word Embeddings (Word2Vec)

1️⃣ Entraînement du modèle Word2Vec

```python
from gensim.models import Word2Vec

sentences = [
    ["fatima", "aime", "ia"],
    ["karim", "aime", "nlp"],
    ["ia", "et", "nlp", "sont", "lies"]
]

model = Word2Vec(
    sentences,
    vector_size=50,
    window=2,
    min_count=1,
    workers=4
)
```

2️⃣ Vecteur associé au mot « aime »

```python
print(model.wv["aime"])
```

✔️ Le résultat est un vecteur numérique représentant le sens du mot

3️⃣ Mots les plus proches de « IA »

```python
print(model.wv.most_similar("ia"))
```

Conclusion:

| Méthode  | Avantages                | Limites                |
| -------- | ------------------------ | ---------------------- |
| BoW      | Simple à implémenter     | Ignore le sens         |
| TF-IDF   | Pondération intelligente | Toujours sans contexte |
| Word2Vec | Capture le sens          | Besoin de données      |
